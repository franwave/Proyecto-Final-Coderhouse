# Opción de algortimo 2: K-vecinos

Debido a la alta presencia de valores nulos en la categoría "Rating"  (casi la mitad), intentaremos predecirlos para completar el dataframe y poder clasificar los juegos mediante una variable más.

```python
# Armaremos dos dataframes (uno únicamente con varibales categoricas y otro con numéricas) para no tener problemas con los tipos de datos
df_categorico = df_limpio[["Platform","Genre","Publisher","Developer","Rating"]]
df_numerico = df_limpio.drop(df_categorico, axis=1)
df_numerico["User_Score"] = df_numerico["User_Score"].astype(float)
```

```python
# Valores presentes en la columna de Rating
df_categorico["Rating"].unique()
```

```python
# Elegimos variables predictoras (X) y variables target (y)
X_KNC = df_categorico.drop("Rating", axis=1)
y_KNC = df_categorico.Rating
```

```python
# Transformación a variables dummies
X_KNC = pd.get_dummies(X_KNC)
y_KNC = pd.get_dummies(y_KNC)
```

```python
# Importación del modelo
from sklearn.neighbors import KNeighborsClassifier
# Creación de modelo
KNC = KNeighborsClassifier(n_neighbors=5)
```

```python
# Partición de entrenamiento (80% train y 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_KNC, y_KNC, test_size=0.2)
# Entrenamiento de modelo
KNC.fit(X_train,y_train)
# Predicción del modelo
y_pred = KNC.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy score del modelo es:', accuracy_score(y_test, y_pred))
```
